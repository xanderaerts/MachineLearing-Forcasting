**TASK: The label we are trying to predict is the SalePrice column. Separate out the data into X features and y labels**

```python
SalePrice = df['SalePrice']
del df['SalePrice']
```

**TASK: Use scikit-learn to split up X and y into a training set and test set. Since we will later be using a Grid Search strategy, set your test proportion to 10%. To get the same data split as the solutions notebook, you can specify random_state = 101**

```python
DF_train, DF_test, SP_train,SP_test = train_test_split(df,SalePrice,test_size=0.1,random_state=101)
```

```python
randomForest = RandomForestClassifier(n_estimators=100,max_features='sqrt')
randomForest.fit(DF_train,SP_train)


RF_preds = randomForest.predict(DF_test)
```

```python
ac = accuracy_score(SP_test,RF_preds)
print(f"{round(ac*100,2)}%")
mae = mean_absolute_error(SP_test,RF_preds)
print(mae)
rmse = root_mean_squared_error(SP_test,RF_preds)
print(rmse)
```
